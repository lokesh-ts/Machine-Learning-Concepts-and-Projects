 ## using scikit-learn to estimate Coefficients

```python
from sklearn.linear_model import LogisticRegression
```

```python
dataset=[[-2.0011,0],
         [-1.4654,0],
         [0.0965,0],
         [1.3881,0],
         [3.0641,0],
         [7.6275,1],
         [5.3324,1],
         [6.9225,1],
         [8.6754,1],
         [7.6737,1]]
```

```python
X=np.array(dataset)[:,0:1]
y=np.array(dataset)[:,1]
```

```python
X
```

```python
y
```

```python
clf_LR=LogisticRegression(C=1,penalty='l2',tol=0.0001)
```

```python
clf_LR.fit(X,y)
```

```python
clf_LR.predict(X)
```

```python
clf_LR.predict_proba(X)
```
